```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


df = pd.read_csv("C:/Users/SABAREESH/Downloads/Placement_Data_Full_Class.csv")

sns.countplot(x='hsc_b', hue='status', data=df)
plt.title('Number of Placed and Not Placed Students by Board of Education')
plt.xlabel('Board of Education')
plt.ylabel('Number of Students')
plt.show()

sns.barplot(x='hsc_b', y='salary', data=df)
plt.title('Average Salary Offered by Board of Education')
plt.xlabel('Board of Education')
plt.ylabel('Salary')
plt.show()

sns.countplot(x='specialisation', hue='status', data=df)
plt.title('Number of Placed and Not Placed Students by MBA Specialization')
plt.xlabel('MBA Specialization')
plt.ylabel('Number of Students')
plt.show()



import seaborn as sns

# Box plot for ssc_b vs salary
sns.boxplot(x='ssc_b', y='salary', data=df)
plt.title('Salary vs SSC Board of Education')
plt.show()

# Box plot for hsc_b vs salary
sns.boxplot(x='hsc_b', y='salary', data=df)
plt.title('Salary vs HSC Board of Education')
plt.show()
```
